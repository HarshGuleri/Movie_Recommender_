```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','budget','vote_average','cast','crew']]
```

```python

movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
keyword_counts = movies['keywords'].explode().value_counts()

# Select the top N keywords for visualization (e.g., top 10)
top_keywords = keyword_counts.head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
top_keywords.plot(kind='bar', color='skyblue')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Top 10 Movie Keywords')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better display
plt.show()
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
genre_counts = movies['genres'].value_counts()
genre_counts.plot(kind='pie')
plt.title('Genre Distribution')
plt.show()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
# CAST POPULARITY
cast_votes = movies['cast'].explode().value_counts()
cast_popularity = cast_votes.groupby(level=0).sum().reset_index(name='vote_count')
cast_popularity['popularity'] = cast_popularity['vote_count'] / cast_popularity['vote_count'].sum()
cast_popularity = cast_popularity.nlargest(10, 'popularity')
cast_popularity.plot(kind='bar', x='cast', y='popularity')
plt.title('Cast Popularity')
plt.show()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Flatten the 'crew' column which contains lists of directors
directors_list = [director for sublist in movies['crew'] for director in sublist]

# Count the occurrences of each director
director_counts = Counter(directors_list)

# Convert the Counter object to a dictionary for plotting
director_counts_dict = dict(director_counts)

# Sort the director counts dictionary by values in descending order
sorted_directors = dict(sorted(director_counts_dict.items(), key=lambda item: item[1], reverse=True))

# Extract top N directors for better visualization
top_directors = dict(list(sorted_directors.items())[:10])  # Adjust the number as needed

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.barh(list(top_directors.keys()), list(top_directors.values()), color='skyblue')
plt.xlabel('Number of Movies Directed')
plt.ylabel('Director')
plt.title('Top Directors by Number of Movies Directed')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.sample(6)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer #count no. of time word appear
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
import matplotlib.pyplot as plt

def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_movies = [new.iloc[i[0]].title for i in distances[1:6]]
    return recommended_movies

# Example usage:
recommended_movies = recommend('The Lego Movie')

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.barh(recommended_movies, [5, 4, 3, 2, 1])  # You can adjust the scores as needed
plt.xlabel('Similarity Score')
plt.title('Top 5 Recommended Movies')
plt.gca().invert_yaxis()  # Invert y-axis for better visualization
plt.show()
```
